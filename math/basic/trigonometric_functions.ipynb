```python
import matplotlib.pyplot as plt
import numpy as np

#   (0 ~ 360)
degrees = np.arange(0, 1440,1)
radians = np.deg2rad(degrees)  #  

#   
sin_values = np.sin(radians)
cos_values = np.cos(radians)
tan_values = np.tan(radians)

# tan       ,    NaN     
tan_values = np.where(np.abs(tan_values) > 10, np.nan, tan_values)

#  
fig, axs = plt.subplots(3, 1, figsize=(12, 10))

# sin
axs[0].plot(degrees, sin_values, label='sin()', linewidth=2)
axs[0].set_title('y = sin()', fontsize=14)
axs[0].set_ylabel('sin()')
axs[0].grid(True)
axs[0].legend()

# cos
axs[1].plot(degrees, cos_values, label='cos()', color='green', linewidth=2)
axs[1].set_title('y = cos()', fontsize=14)
axs[1].set_ylabel('cos()')
axs[1].grid(True)
axs[1].legend()

# tan
axs[2].plot(degrees, tan_values, label='tan()', color='red', linewidth=2)
axs[2].set_title('y = tan()', fontsize=14)
axs[2].set_xlabel(' (degrees)')
axs[2].set_ylabel('tan()')
axs[2].set_ylim(-10, 10)
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```
