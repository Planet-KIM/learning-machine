```python
import matplotlib.pyplot as plt
import numpy as np

# 각도 범위 (0도 ~ 360도)
degrees = np.arange(0, 1440,1)
radians = np.deg2rad(degrees)  # 라디안 변환

# 함수 값 계산
sin_values = np.sin(radians)
cos_values = np.cos(radians)
tan_values = np.tan(radians)

# tan 함수는 수직 점근선 근처에서 값이 무한대로 커지므로, 너무 큰 값은 NaN으로 처리해 그래프가 튀지 않게 함
tan_values = np.where(np.abs(tan_values) > 10, np.nan, tan_values)

# 그래프 그리기
fig, axs = plt.subplots(3, 1, figsize=(12, 10))

# sin
axs[0].plot(degrees, sin_values, label='sin(θ)', linewidth=2)
axs[0].set_title('y = sin(θ)', fontsize=14)
axs[0].set_ylabel('sin(θ)')
axs[0].grid(True)
axs[0].legend()

# cos
axs[1].plot(degrees, cos_values, label='cos(θ)', color='green', linewidth=2)
axs[1].set_title('y = cos(θ)', fontsize=14)
axs[1].set_ylabel('cos(θ)')
axs[1].grid(True)
axs[1].legend()

# tan
axs[2].plot(degrees, tan_values, label='tan(θ)', color='red', linewidth=2)
axs[2].set_title('y = tan(θ)', fontsize=14)
axs[2].set_xlabel('θ (degrees)')
axs[2].set_ylabel('tan(θ)')
axs[2].set_ylim(-10, 10)
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```
